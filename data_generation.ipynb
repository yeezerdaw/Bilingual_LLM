```python
import os, json, re

sample_file = "data/tamil_wikipedia/AA/wiki_00"

def clean_wiki_text(text):
    # Remove HTML entities
    text = re.sub(r"&lt;.*?&gt;", " ", text)
    # Remove section headers like "== something =="
    text = re.sub(r"={2,}.*?={2,}", " ", text)
    # Remove brackets, templates, etc.
    text = re.sub(r"\{\{.*?\}\}", " ", text)
    # Remove URLs
    text = re.sub(r"http\S+", " ", text)
    # Collapse whitespace
    text = re.sub(r"\s+", " ", text)
    return text.strip()

cleaned_articles = []

with open(sample_file, "r") as f:
    for i, line in enumerate(f):
        if i >= 5:
            break
        article = json.loads(line)
        text = clean_wiki_text(article["text"])
        if len(text) > 100:  # skip tiny or empty
            cleaned_articles.append({
                "title": article["title"],
                "text": text[:400] + "..."  # truncate preview
            })

print(json.dumps(cleaned_articles, indent=2, ensure_ascii=False))
```

```python
import os, json, re
from tqdm import tqdm

base_dir = "data/tamil_wikipedia"
output_path = "data_cleaned/tamil_wikipedia_clean.jsonl"
os.makedirs("data_cleaned", exist_ok=True)

def clean_wiki_text(text):
    text = re.sub(r"&lt;.*?&gt;", " ", text)
    text = re.sub(r"={2,}.*?={2,}", " ", text)
    text = re.sub(r"\{\{.*?\}\}", " ", text)
    text = re.sub(r"http\S+", " ", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()

with open(output_path, "w") as out_f:
    for root, dirs, files in os.walk(base_dir):
        for fname in files:
            if not fname.startswith("wiki_"): 
                continue
            fpath = os.path.join(root, fname)
            with open(fpath, "r") as f:
                for line in f:
                    article = json.loads(line)
                    text = clean_wiki_text(article["text"])
                    if len(text) > 100:  # skip stubs / templates
                        json.dump(
                            {"title": article["title"], "text": text},
                            out_f,
                            ensure_ascii=False
                        )
                        out_f.write("\n")

print(f"✅ Tamil Wikipedia cleaned corpus saved to: {output_path}")
```
